### The Diachronic Interpretation <a id="diachronic"></a>

So far, we have learned about how we can use Bayes' theorem in the presence of uncertainty to solve conditional probability problems.

A real-world application of this approach allows us to:

**update the probability of a hypothesis, _H_, in light of some body of data, _D_.**

Rewriting Bayes' theorem with Hypothesis and Data yields:

* $P(H)$ is the probability of the hypothesis _H_ before we see the data, called the **prior probability**

* $P(H\, |\, D)$ is the probability of the hypothesis H after we see the data D, called the **posterior probability**

* $P(D\,|\, H)$ is the probability of the data D under hypothesis H, called the **likelihood**

* $P(D)$ is the probability of the data D under any hypothesis, called the **evidence** or **normalizing constant**

How do we choose the prior?

The prior represents our personal perspective, our beliefs, and our uncertainties.

We are representing a cumulative distribution for the parameters such that:

* it is practical
* in practice we work with a family of functions
* we build from previous experiments
* if there is a large amount of data, then most reasonable priors will give similar results

#### Example: The M & M problem

In 1995, blue M & M’s were introduced.

Before then, the color mix was 30% brown, 20% yellow, 20 % red, 10% green, 10% orange, 10% tan.

Afterward, it was 24% blue, 20 % green, 16% orange, 14% yellow, 13% red, 13% brown.

You have 2 bags: one is from 1994, and one from 1996. You draw one M & M from each bag. One is yellow and one is green.

* What is the probability that the yellow one came from the 1994 bag?

**Hypothesis A**: Yellow is from the 1994 bag, and Green is from the 1996 bag.

**Hypothesis B**: Yellow is from the 1996 bag, and Green is from the 1994 bag.

What is a reasonable *prior*? ($\frac{1}{2}$)

Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
:---:|:---:|:---:|:---:|:---:|
A|$\frac{1}{2}$|-|-|-|
B|$\frac{1}{2}$|-|-|-|

Given Hypothesis A, what is the *likelihood* of Yellow/Green? $(0.2 \times 0.2)$

Given Hypothesis B, what is the *likelihood* of Yellow/Green? $(0.14 \times 0.10)$

Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
:---:|:---:|:---:|:---:|:---:|
A|$\frac{1}{2}$|$0.2 \times 0.2$|-|-|
B|$\frac{1}{2}$|$0.14 \times 0.10$|-|-|

We can multiply the prior and the likelihood to get the numerator, and, since these are the only options, we can use their sum for the denominator (0.027).

This process is summarized in the table below:

Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
:---:|:---:|:---:|:---:|:---:|
A|$\frac{1}{2}$|$0.2 \times 0.2$|$0.2 \times 0.2 \times 0.5 = 0.02$|$0.02\ /\ 0.027 = 0.74$|
B|$\frac{1}{2}$|$0.14 \times 0.10$|$0.14 \times 0.1 \times 0.5 = 0.007$|$0.007\ /\ 0.027 = 0.26$|

**Thus, the probability that the yellow M & M came from the 1994 bag (Hypothesis A) is 0.74.**

#### Example: The Monty Hall Problem


You are presented with three closed doors: behind one is a car, behind each of the other two is a goat. If you choose the door with the car behind it, you keep the car.

* Step 1: You pick a door (Example – **Door A** – still unknown)
* Step 2: The host opens one of the doors you didn’t choose, revealing a goat (Example – **Door B** – goat)
* Step 3: You are given the option to stick with **Door A**, or switch to **Door C**.

QUESTION: Do you stick with **Door A** or switch to **Door C**?

You might think that it doesn't matter! Isn't the result going to be the same? **But it does.**

#### Bayesian Solution to the Monty Hall Problem

![montyhall1.png](attachment:montyhall1.png "Monty Hall")

Assuming that the prizes are distributed randomly, what is a reasonable *prior*? ($\frac{1}{3}$)

Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
:---:|:---:|:---:|:---:|:---:|
A|$\frac{1}{3}$|-|-|-|
B|$\frac{1}{3}$|-|-|-|
C|$\frac{1}{3}$|-|-|-|

Our choice was **Door A**, and Monty opened **Door B**. What is the *likelihood* of Monty opening **Door B** given that we picked **Door A** in terms of the three hypotheses?

(Remember that we are trying to determine the probability of the obtained observations given each hypothesis.)

* If the car were behind **Door A**, then Monty could have opened **Door B** or **Door C**; assuming this is random,
there is a 50% chance of either event happening.
* If the car were behind **Door B**, then Monty could not have opened this door, so there is no chance that this is the case.
* If the car were behind **Door C**, then things could only have played out the way they have (as he can’t open your selected door)

Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
:---:|:---:|:---:|:---:|:---:|
A|$\frac{1}{3}$|$\frac{1}{2}$|-|-|
B|$\frac{1}{3}$|0|-|-|
C|$\frac{1}{3}$|1|-|-|

Multiplying this out to get $P(D|H)P(H)$:

Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
:---:|:---:|:---:|:---:|:---:|
A|$\frac{1}{3}$|$\frac{1}{2}$|$\frac{1}{6} = 0.17$|-|
B|$\frac{1}{3}$|$0$|$\ \ \ \ \ \ \ \ \ \ \ \ \ 0\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ $|-|
C|$\frac{1}{3}$|$1$|$\frac{1}{3} = 0.33$|-|

Again, there were only these three scenarios, so we can normalize by the sum of the probabilities:

|Hypothesis $$H$$|Prior $$P(H)$$|Likelihood $$P(D \mid H)$$|Likelihood * Prior$$P(D \mid H)P(H)$$|Posterior $$P(H \mid D)$$|
|:---:|:---:|:---:|:---:|---:|
|A|$\frac{1}{3}$|$\frac{1}{2}$|$\frac{1}{6} = 0.17$|$0.167\ /\ 0.500 = \textbf{0.333}$|
|B|$\frac{1}{3}$|$0$|$\ \ \ \ \ \ \ \ \ \ \ \ \ 0\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ $|$0$|
|C|$\frac{1}{3}$|$1$|$\frac{1}{3} = 0.33$|$0.333\ /\ 0.500 = \textbf{0.667}$|

It turns out you are **twice as likely** to be successful if you **switch** your selection! This is not an intuitive result. But the following table should convince you:

Car|Choice|Win if Switch|Win if Stick|
:---:|:---:|:---:|:---:|
A|A|No|Yes
A|B|**Yes**|No
A|C|**Yes**|No
B|A|**Yes**|No
B|B|No|**Yes**
B|C|**Yes**|No
C|A|**Yes**|No
C|B|**Yes**|No
C|C|No|**Yes**

You can see that you win if you **switch** 6 / 9 times, and if you **stick** only 3 / 9 times.
